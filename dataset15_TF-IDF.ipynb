```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

tfid = TfidfVectorizer(stop_words={''})

doc = ["The burger is tasty", "The food is delicious", "The food is awful"]

doc_vector = tfid.fit_transform(doc)
print(tfid.get_feature_names())
df=pd.DataFrame(doc_vector.todense(),columns=tfid.get_feature_names())
print(df)
#print(doc_vector)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

header_names=['text']
df = pd.read_csv('descriptions.csv', header=None, skiprows=1, names=header_names, sep='\n')

tfid = TfidfVectorizer(stop_words={''})

doc_vector = tfid.fit_transform(df['text']) # MEGET STORT OUTPUT
print(tfid.get_feature_names())
df=pd.DataFrame(doc_vector.todense(),columns=tfid.get_feature_names())
print(df)



```
